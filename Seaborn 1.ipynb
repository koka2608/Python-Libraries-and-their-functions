# Seaborn

Seaborn is a statistical plotting library

It is built on top of matplotplot lib

Designed to work on pandas datframe object


```python
import seaborn as sns

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Seaborn actually comes with a few built in data sets that we can actually load

# We will be using that

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Dist plot

Dist plot allows us to show the univariate set of observations

Univariate means one variable

```python
# Plotting distplot

sns.distplot(tips['total_bill'])

#The line we get is KDE (Kernel density estimation)
```

```python
#We can remove KDE using KDE = false

sns.distplot(tips['total_bill'], kde = False, bins = 30)  #distplot is like histogram
```

```python
# Joint plot

# Joint plot allows you to combine bivariate data. It means it can essentially allow to combine 2 distplot
# It means comparing 2 variables in the data

# Syntax 
#    sns.jointplot(x = 'column_name', y = 'column_name', data = tips)

#x and y are the variables that we want to compare

# we have one more parameter in distplot which is called as kind
# kind actually controls what is going on inside the plot i.e by default it is scatter plot
# kind = 'hex' allows to make hexagonal distribution, we can even put as reg which is regression

sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = 'hex')

sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = 'reg')

sns.jointplot(x = 'total_bill', y = 'tip' , data = tips, kind = 'kde')

sns.jointplot(x = 'total_bill', y = 'tip' , data = tips)

```

Next is Pair Plot

```python
# Pairplot() is going to plot pairwise realtionship across an entire data frame
# pair plot is going to joinplot for every single combination of the numerical columns in the data frame

# Syntax :
#    pairplot(dataframe_name)     larger the df more time the pair plot will take

# it is a nice way to quickly visualize your data

sns.pairplot(tips, hue = 'sex')

#using hue it will color code the categorical variable

```

```python

sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
# Now we will use rugplot
 # same like distplot just pass in a column of the df here

sns.rugplot(tips['total_bill'])
```

# Categorical plot

```python
tips.head()
```

```python
# basic categorical plot is barplot
# Categorical data is the data which is divided in categories for example in our dataframe,
# sex, smoker and time

# Syntax : 
    # sns.barplot(x = categorical_column, y = numerical_column, data = data_frame)


sns.barplot(x = 'sex', y = 'total_bill', data = tips)

# We can add an estimator as an argument which is basically an aggregate function. by default it is avg or mean

sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)

```

```python
# Countplot
# Countplot is like a barplot but it just counts the no of observations

sns.countplot(x = 'sex' , data = tips)

#It just counts the data points
```

```python
# Box plot
# It shows the distribution of quantitative data which shows comparison between variables

sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')

# hue allows us to categorize them more into categoraical columns
```

```python
# Violinplot

# It is just like a box plot

sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)

#Split = true will split the plot into 2 and each categorical plot will have their own distribution
```

```python
sns.stripplot(x = 'sex', y = 'total_bill', data = tips, jitter = True, hue = 'sex', split = True)

# Jitter adds more noise to separate some of the stacked points
```
