```python
# We can observe in our dataset that we have few names with mega in them
# Suppose we want to filter them out we can do the following

# Syntax : .str.contains('String_name')

df.loc[df['Name'].str.contains('Mega')]


# We can see that all of these are the columns that have mega in them
```

```python
# We can also use the not function to get values which do not have mega in them

#  The symbol ' ~ ' is used to drop those lines


df.loc[~df['Name'].str.contains('Mega')]
```

```python
# We can also filter out values based on regex 

import re

df.loc[df['Type 1'].str.contains('Fire|Grass', regex = True)]

# Note fire and Grass should contain the proper string
```

```python
# We can use flag to avoid that

df.loc[df['Type 1'].str.contains('fire|grass', regex = True, flags = re.I)]

# We will get the same as above
```

```python
# Get all pokemon names starting with 'PI'


df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex = True)]
```

# Aggregate Statistic i.e Group by

```python
# Loading the csv

df = pd.read_csv('modified.csv', delimiter = '\t')
```

```python
df
```

```python
# We can group by pokemons by their Skills like we do in SQL

# groupby() function is used to do that

# sort_values() function is used to sort them by a particular column. It is like a Order by in SQL

df.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)  



# here we get all stats by their means and sorted as what type they are

# Steel pokemon have the highest defense
```

```python
# lets see who has the highest attack

df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)  


# Pokemon of type Dragon has highest attack
```

```python
# Let's see who has best speed

df.groupby(['Type 1']).mean().sort_values('Speed', ascending = False)  
```

```python
# We can also do sum()

df.groupby(df['Type 1']).sum()


# We will have all HP's added up. Does not make sense
```

```python
# We can also do count

df.groupby(df['Type 1']).count().sort_values('HP', ascending = False)  # Counting in descending order


# We will get count of each type of pokemon
```

```python
df['count'] = 1

df.groupby(['Type 1']).count()['count']
```

```python
# We can also group by by using 2 columns

df.groupby(['Type 1', 'Type 2']).count()['count']
```
