```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Load in 2 default seaborn data sets

tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

tips.head()
```

```python
flights.head()
```

```python
#Heatmap 

# Step 1: Convert the data frame in matrix form
# 
# df_name.pivot_table(index = 'column_name', columns = 'column_name', values = 'column_name')
# majorly index and columns have nominal data 

f = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')

#Check the pivot_table
f
```

```python
# Step 2:
#    Call the heatmap function

sns.heatmap(f)
```

```python
#Customizing the heatmap some more

sns.heatmap(f, linecolor = 'white', linewidth = 1)
```

```python
sns.clustermap(f, standard_scale= 1)
```

Now we will create grids using seaborn

```python
iris = sns.load_dataset('iris')

iris.head()
```

```python
sns.pairplot(iris)
```

```python
#Create a variable and assign the pairgrid to a variable

g = sns.PairGrid(iris)

#Then map the whole variable plot by plot
# USe the map function for that

g.map_diag(sns.distplot)

g.map_upper(plt.scatter)

g.map_lower(sns.kdeplot)
```
